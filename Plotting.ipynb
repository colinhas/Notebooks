```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl


data = np.empty((96, 4))
with open(r"C:\Users\squid\Downloads\Well_treatments 0.75 and 1.5%_CFUs.xlsx - Sheet1 (1).csv", 'r')as file:
    filecontent=csv.reader(file)
    i = 0
    for row in filecontent:
        if i != 0:
            data[i-1, 0] = row[8]
            data[i-1, 1] = row[3]
            data[i-1, 2] = row[6]
            data[i-1, 3] = row[9]
        #print(row)
        i += 1
print(data)
```
