```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl


data = np.empty((96, 4))
with open(r"C:\Users\squid\Downloads\Well_treatments 0.75 and 1.5%_CFUs.xlsx - Sheet1 (1).csv", 'r')as file:
    filecontent=csv.reader(file)
    i = 0
    for row in filecontent:
        if i != 0:
            data[i-1, 0] = row[8]
            data[i-1, 1] = row[3]
            data[i-1, 2] = row[6]
            data[i-1, 3] = row[9]
        #print(row)
        i += 1
print(data)
```

```python
dat_1 = data[data[:,3] ==1,:]

print(dat_1)
```

```python
dat_2 = data[data[:,3] ==2,:]

print(dat_2)
```

```python
dat_3 = data[data[:,3] ==3,:]

print(dat_3)
```

```python
dat_4 = data[data[:,3] ==4,:]

print(dat_4)
```

```python
dat_5 = data[data[:,3] ==5,:]

print(dat_5)
```

```python
dat_6 = data[data[:,3] ==6,:]

print(dat_6)
```

```python
fig, ax = plt.subplots(figsize=(10, 5), layout='constrained')
ax.plot(dat_1[:,1],dat_1[:,0], label = "Carb")
ax.plot(dat_2[:,1],dat_2[:,0], label = "RIF")
ax.plot(dat_3[:,1],dat_3[:,0], label = "PmB")
ax.plot(dat_4[:,1],dat_4[:,0], label = "CAM")
ax.plot(dat_5[:,1],dat_5[:,0], label = "Cipro")
ax.plot(dat_6[:,1],dat_6[:,0], label = "KAN")

plt.gca().invert_xaxis()

ax.set_title('Concentration of Drug vs Number of CFUs')
ax.set_ylabel('CFUs')
ax.set_xlabel('Drug Conc (ug/mL)')
ax.grid()
ax.legend();
```
