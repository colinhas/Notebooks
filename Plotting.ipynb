```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl


data = np.empty((96, 4))
with open(r"C:\Users\squid\Downloads\Well_treatments 0.75 and 1.5%_CFUs.xlsx - Sheet1 (1).csv", 'r')as file:
    filecontent=csv.reader(file)
    i = 0
    for row in filecontent:
        if i != 0:
            data[i-1, 0] = row[8]
            data[i-1, 1] = row[3]
            data[i-1, 2] = row[6]
            data[i-1, 3] = row[9]
        #print(row)
        i += 1
print(data)
```

```python
dat_1 = data[data[:,3] ==1,:]
dat_1[dat_1[:,0] ==0,0] =0.1
dat_1_15 = dat_1[dat_1[:,2] ==1.5,:]
print(dat_1_15)
```

```python
dat_2 = data[data[:,3] ==2,:]
dat_2[dat_2[:,0] ==0,0] =0.1
dat_2_15 = dat_2[dat_2[:,2] ==1.5,:]

print(dat_2_15)
```

```python
dat_3 = data[data[:,3] ==3,:]
dat_3[dat_3[:,0] ==0,0] =0.1
dat_3_15 = dat_3[dat_3[:,2] ==1.5,:]

print(dat_3_15)
```

```python
dat_4 = data[data[:,3] ==4,:]
dat_4[dat_4[:,0] ==0,0] =0.1
dat_4_15 = dat_4[dat_4[:,2] ==1.5,:]

print(dat_4_15)
```

```python
dat_5 = data[data[:,3] ==5,:]
dat_5[dat_5[:,0] ==0,0] =0.1
dat_5_15 = dat_5[dat_5[:,2] ==1.5,:]

print(dat_5_15)
```

```python
dat_6 = data[data[:,3] ==6,:]
dat_6[dat_6[:,0] ==0,0] =0.1
dat_6_15 = dat_6[dat_6[:,2] ==1.5,:]

print(dat_6_15)
```

```python
fig, ax = plt.subplots(figsize=(10, 5), layout='constrained')
ax.semilogy(dat_1_15[:,1]/20,dat_1_15[:,0], label = "Carb")
ax.semilogy(dat_2_15[:,1]/20,dat_2_15[:,0], label = "RIF")
ax.semilogy(dat_3_15[:,1]/20,dat_3_15[:,0], label = "PmB")
ax.semilogy(dat_4_15[:,1]/20,dat_4_15[:,0], label = "CAM")
ax.semilogy(dat_5_15[:,1]/20,dat_5_15[:,0], label = "Cipro")
ax.semilogy(dat_6_15[:,1]/20,dat_6_15[:,0], label = "KAN")

plt.gca().invert_xaxis()

ax.set_title('Concentration of Drug vs Number of CFUs at 1.5% Agarose')
ax.set_ylabel('CFUs')
ax.set_xlabel('Drug Conc (ug/mL)')
ax.grid()
ax.legend();
```
